```python
a1, b1 = map(int, input().split())
a2, b2 = map(int, input().split())

import math
bb = math.lcm(b1, b2)
aa1 = a1 * (bb//b1)
aa2 = a2 * (bb//b2)
aa12 = aa1+aa2

i = math.gcd((aa12), bb)

# ê¸°ì•½ë¶„ìˆ˜ë¥¼ ë§Œë“¤ì–´ì•¼ í•¨!!
if i == 1 :
    print(aa12, bb)
else :
    aa12 //= i
    bb //= i
    print(aa12, bb)
```

- **ì…ë ¥** : ì²«ì§¸ ì¤„ì—ëŠ” ì´ë¯¸ ì‹¬ì–´ì ¸ ìˆëŠ” ê°€ë¡œìˆ˜ì˜ ìˆ˜ë¥¼ ë‚˜íƒ€ë‚´ëŠ” í•˜ë‚˜ì˜ ì •ìˆ˜ Nì´ ì£¼ì–´ì§„ë‹¤(3 â‰¤ N â‰¤ 100,000). ë‘˜ì§¸ ì¤„ë¶€í„° Nê°œì˜ ì¤„ì—ëŠ” ê° ì¤„ë§ˆë‹¤ ì‹¬ì–´ì ¸ ìˆëŠ” ê°€ë¡œìˆ˜ì˜ ìœ„ì¹˜ê°€ ì–‘ì˜ ì •ìˆ˜ë¡œ ì£¼ì–´ì§€ë©°, ê°€ë¡œìˆ˜ì˜ ìœ„ì¹˜ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ì •ìˆ˜ëŠ” 1,000,000,000 ì´í•˜ì´ë‹¤. ê°€ë¡œìˆ˜ì˜ ìœ„ì¹˜ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ì •ìˆ˜ëŠ” ëª¨ë‘ ë‹¤ë¥´ë‹¤.
 
- **ì¶œë ¥** : ëª¨ë“  ê°€ë¡œìˆ˜ê°€ ê°™ì€ ê°„ê²©ì´ ë˜ë„ë¡ ìƒˆë¡œ ì‹¬ì–´ì•¼ í•˜ëŠ” ê°€ë¡œìˆ˜ì˜ ìµœì†Œìˆ˜ë¥¼ ì²« ë²ˆì§¸ ì¤„ì— ì¶œë ¥í•œë‹¤.

- **ì˜ˆì œ ì…ì¶œë ¥** 
    - ì…ë ¥1
        - `4`
        - `1`
        - `3`
        - `7`
        - `13`
        
    - ì¶œë ¥1
        - `3`

```python
# ì‹œê°„ ì´ˆê³¼....
n = int(input())
tree = []
for _ in range(n) :
    tree.append(int(input()))

# tree listì—ì„œ ë‚˜ë¬´ê°€ ì‹¬ì–´ì ¸ ìˆëŠ” ê°„ê²© ì°¾ê¸°
s = []
for i in range(n-1) :
    s.append(tree[i+1]-tree[i])
    
# s listì—ì„œ ìµœëŒ€ê³µì•½ìˆ˜ ì°¾ê¸°
import math
gcd = s[0]
for item in s :
    gcd = math.gcd(gcd, item)

# ì›ë˜ ì‹¬ì–´ì ¸ ìˆëŠ” ë‚˜ë¬´ ê°œìˆ˜
origin_n = len(tree)

# ë‚˜ë¬´ ì¶”ê°€
g = tree[0]
while True :
    g += gcd
    if g not in tree :
        tree.append(g)
    tree.sort()
    if g==tree[-1] :
        break

print(len(tree) - origin_n)
```

```python
# ìµœëŒ€í•œ forë¬¸ì„ ì•ˆì¨ì•¼í•¨
from math import gcd
n = int(input())
a = int(input())

tree = []

for _ in range(n-1) :
    num = int(input())
    tree.append(num-a)
    a = num
    
g = tree[0]
for t in tree :
    g = gcd(t, g)
    
# key point !!!!
result = 0
for tr in tree :
    result += tr//g - 1
    
print(result)
```

```python
# ì‹œê°„ì´ˆê³¼...
import math
import sys

n = int(sys.stdin.readline())
for _ in range(n) :
    nn = int(sys.stdin.readline())
    while True :
        y = 0
        for i in range(2, int(math.sqrt(nn))+1) :
            if nn%i == 0 :
                y += 1
        if y == 1 :
            print(nn)
            break
    
        else :
            nn += 1
```

```python
# ì‹œê°„ ì´ˆê³¼
import math
import sys
def sosu(x) :
    for i in range(2, int(math.sqrt(x))+1) :
        if x%i == 0 :
            return False
    return True

t = int(sys.stdin.readline())
for _ in range(t) :
    n = int(sys.stdin.readline())
    sosu_n = 0
    for i in range(2, n//2+1) :
        if sosu(i) == True :
            if sosu(n-i) == True :
                sosu_n += 1
    print(sosu_n)
```

```python
# ë©”ëª¨ë¦¬ ì´ˆê³¼...
n = int(input())

chang = [0 for i in range(n+1)]

for i in range(1, n+1, 1) :
    # ë¨¼ì € ì²«ë²ˆì§¸ ì‚¬ëŒ -> 1ì˜ ë°°ìˆ˜ ì°½ë¬¸ êµì²´
    for j in range(i, n+1, i) :
        if chang[j] == 0 :
            chang[j] = 1
        elif chang[j] == 1 :
            chang[j] = 0
            
total = 0
for c in chang :
    total += c
print(total)
```
