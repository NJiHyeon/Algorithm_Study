```python
# ëŸ°íƒ€ì„ ì—ëŸ¬..
t = int(input())
for i in range(t) :
    a, b = map(int, input().split())
    
    if a > b :
        e = 45000//b
        a_list = []
        b_list = []
        for i in range(1, e+1) :
            a_list.append(a*i)
            b_list.append(b*i)
        
        ab = Counter(a_list + b_list)
    
        ab_set = list(set(a_list + b_list))
        ab_set.sort()
    
        for j in ab_set :
            if ab[j] == 2 :
                print(j)
                break
            
    elif a < b :
        e = 45000//a
        a_list = []
        b_list = []
        for i in range(1, e+1) :
            a_list.append(a*i)
            b_list.append(b*i)
        
        ab = Counter(a_list + b_list)
    
        ab_set = list(set(a_list + b_list))
        ab_set.sort()
    
        for j in ab_set :
            if ab[j] == 2 :
                print(j)
                break
            
    elif a==b :
        print(a)
```

#### âœ”ï¸ë‹¨ê³„3 : 1735ë²ˆ(ë¶„ìˆ˜ í•©)

- **ë¬¸ì œ** : ë¶„ìˆ˜ A/BëŠ” ë¶„ìê°€ A, ë¶„ëª¨ê°€ Bì¸ ë¶„ìˆ˜ë¥¼ ì˜ë¯¸í•œë‹¤. Aì™€ BëŠ” ëª¨ë‘ ìì—°ìˆ˜ë¼ê³  í•˜ì.


- **ì…ë ¥** : ë‘ ë¶„ìˆ˜ì˜ í•© ë˜í•œ ë¶„ìˆ˜ë¡œ í‘œí˜„í•  ìˆ˜ ìˆë‹¤. ë‘ ë¶„ìˆ˜ê°€ ì£¼ì–´ì¡Œì„ ë•Œ, ê·¸ í•©ì„ ê¸°ì•½ë¶„ìˆ˜ì˜ í˜•íƒœë¡œ êµ¬í•˜ëŠ” í”„ë¡œê·¸ë¨ì„ ì‘ì„±í•˜ì‹œì˜¤. ê¸°ì•½ë¶„ìˆ˜ë€ ë” ì´ìƒ ì•½ë¶„ë˜ì§€ ì•ŠëŠ” ë¶„ìˆ˜ë¥¼ ì˜ë¯¸í•œë‹¤.
 
- **ì¶œë ¥** : ì²«ì§¸ ì¤„ê³¼ ë‘˜ì§¸ ì¤„ì—, ê° ë¶„ìˆ˜ì˜ ë¶„ìì™€ ë¶„ëª¨ë¥¼ ëœ»í•˜ëŠ” ë‘ ê°œì˜ ìì—°ìˆ˜ê°€ ìˆœì„œëŒ€ë¡œ ì£¼ì–´ì§„ë‹¤. ì…ë ¥ë˜ëŠ” ë„¤ ìì—°ìˆ˜ëŠ” ëª¨ë‘ 30,000 ì´í•˜ì´ë‹¤.

- **ì˜ˆì œ ì…ì¶œë ¥** 
    - ì…ë ¥1
        - `2 7`
        - `3 5`
        
    - ì¶œë ¥1
        - `31 35`

```python
a1, b1 = map(int, input().split())
a2, b2 = map(int, input().split())

import math
bb = math.lcm(b1, b2)
aa1 = a1 * (bb//b1)
aa2 = a2 * (bb//b2)
aa12 = aa1+aa2

i = math.gcd((aa12), bb)

# ê¸°ì•½ë¶„ìˆ˜ë¥¼ ë§Œë“¤ì–´ì•¼ í•¨!!
if i == 1 :
    print(aa12, bb)
else :
    aa12 //= i
    bb //= i
    print(aa12, bb)
```
