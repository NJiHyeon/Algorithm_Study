## 🔎재귀함수

#### 재귀함수
- 재귀함수란 자기 자신을 다시 호출하는 함수로, DFS를 실질적으로 구현하고자 할 때 자주 사용되는 방법 중 하나이다.
- 아래와 같은 함수를 호출하면 어느 정도 출력하다가 최대 재귀 깊이 초과 메세지가 출력된다.

#### 유의사항
- 재귀함수는 수학적으로 정의된 점화식과 같은 형태를 간단히 코드로 옮길 수 있다는 점에서 매우 직관적이고, 짧은 코드로 함수를 작성할 수 있다.
- 모든 재귀함수는 반복문을 이용하여 동일한 기능을 구현할 수 있다.
- 재귀함수가 반복문보다 유리한 경우도 있고, 불리한 경우도 있다.
- 스택을 사용해야 할 때 구현상 스택 라이브러리 대신에 재귀함수를 이용하는 경우가 많다.

```python
'''
def 함수이름(입력) :
    if 종료 조건 :
        종료 선언
    return 함수이름'(입력')
    print('---')
'''
```

```python
def recursive_function() :
    print('재귀 함수를 호출합니다.')
    recursive_function()
    
recursive_function()
```

```python
def a(i) :
    if a==3 :
        return
    print('3')
```

```python
a(100)
```

```python
a(3)
```

#### 재귀함수 종료 조건
- 재귀 함수를 문제 풀이에서 사용할 때는 재귀 함수의 종료 조건을 반드시 명시해야 한다.
- 종료 조건을 제대로 명시하지 않으면 함수가 무한히 출력될 수 있다.
- **stack에 대한 정보를 넣었다가 꺼내는 것과 마찬가지로, 함수에 대한 정보가 stack frame에 담기게 되어서 차례대로 호출되었다가 -> 가장 마지막에 호출된 함수부터 종료되어 첫번째 호출했던 시점의 함수까지 종료된다.**

```python
def recursive_function(i) :
    if i==100 :
        return
    print(i, '번째 재귀함수에서', i+1, '번째 재귀함수를 호출합니다.')
    recursive_function(i+1)
    print(i, '번째 재귀함수를 종료합니다.')
```

```python
recursive_function(1)
```

#### 팩토리얼 구현 예제

```python
def factorial_recursive(n) :
    if n <= 1 :
        return 1
    return factorial_recursive(n-1) * n
```

```python
factorial_recursive(3)
```

#### 최대공약수 계산 예제 (유클리드 호제법)

```python
def gcd(a, b) :
    if a%b == 0 :
        return b
    return gcd(b, a%b)
```

```python
gcd(192, 162)
```
