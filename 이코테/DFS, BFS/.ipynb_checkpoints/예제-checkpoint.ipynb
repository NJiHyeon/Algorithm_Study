```python
'''
- dfs í•¨ìˆ˜ëŠ” ì§€ì •ëœ ìœ„ì¹˜ì—ì„œ(ì´ˆê¸°ê°’) ë°˜ë³µì ìœ¼ë¡œ ì¬ê·€í•˜ë©´ì„œ í™•ì¸ !
- forë¬¸ì€ ì „ì²´ ë°ì´í„°ì— ëŒ€í•´ í™•ì¸
- ë‘ ê³¼ì •ì„ ìˆ˜í–‰í•˜ë©´ ê²¹ì¹  ìˆ˜ ìˆëŠ” ê°’ì´ ì¡´ì¬í•˜ëŠ”ë° ì´ê²ƒì„ return True/Falseë¡œ ì²˜ë¦¬ !
  - í•œë²ˆ ë‹¤ ìˆ˜í–‰í•˜ë©´ return True
'''
```

```python
n, m = map(int, input().split())
```

```python
drink = []
for i in range(n) :
    l = list(map(int, input()))
    drink.append(l)
```

```python
drink[1][0]
```

```python
def dfs(x, y) :
    # ë§Œì•½ i, jì˜ ìœ„ì¹˜ (í˜„ì¬ ìœ„ì¹˜)ê°€ ë²”ìœ„ë¥¼ ë²—ì–´ë‚˜ì§€ ì•Šìœ¼ë©´ ê³„ì† ìˆ˜í–‰
    if x<=-1 or x >= n or y<=-1 or y>=m :
        return False
    
    # ìƒ, í•˜, ì¢Œ, ìš° ì¬ê·€í•¨ìˆ˜ ëŒë¦¬ê¸°
    # ë°©ë¬¸ì²˜ë¦¬ í•„ìˆ˜ !!
    if drink[x][y] == 0 :
        drink[x][y] = 1
        dfs(x-1, y)
        dfs(x+1, y)
        dfs(x, y-1)
        dfs(x, y+1)
        return True
    #return False
```

```python
ice = 0
for i in range(n) :
    for j in range(m) :
        if dfs(i, j) == True :
            ice += 1
            
print(ice)
```
